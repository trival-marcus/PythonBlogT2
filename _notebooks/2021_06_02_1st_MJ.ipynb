### ③ 매개변수 *args
매개변수가 가변적일 떄 사용한다.
- args 라는 표현은 관용적으로 쓰는 표현.
- 그냥 *a, *aaeerer 등 멋대로 선언해도됨.

````python
# *args, 매개변수. 매개변수의 개수가 가변적일 때 사용.
def average(*args):
    avrg = sum(args) / len(args)
    return avrg

print(average(1,2,3))
print(average(1,2,3,4,5))
````

### 예제1. 소수 여부 판단하기.

````python
# 소수는 1과 자기자신만으로 나누어 떨어지는 1보다 큰 양의 정수다.
# 이를 판단하는 것을 함수 안에 넣어보기

# 1부터 자기자신까지의 범위에서 나누었을 때 0이면 안됨.

def is_prime(number):
    # 소수여부 판단하여 리턴
    # 소수:True, 소수가 아님:False

    if number < 2:
        return False
    
    for i in range(2,number) :
        if number % i == 0:
            return False
        
    return True

input_number = int(input('숫자: '))
if is_prime(input_number):
    print('소수입니다.')
else :
    print('소수가 아닙니다.')
````

````python
# 코드메이트님 코드

def is_prime(num):
    #소수여부 
    #소수 True ,아니면 False
    if num < 2:
        return False
    for i in range(2,num):
        if num % i == 0:
            return False
    return num

def print_prime(num):
    prime_list = []
    for i in range(num):
        if is_prime(i):
            prime_list.append(i)
    
    return prime_list

print_prime(int(input("범위 입력: ")))
````

### ② map(함수명, 순서있는 자료형)

> 들여쓴 블록


두개의 인자를 갖는 함수. 순서가 있는 자료형에 사용.

두 인자를 매칭시킬 때 쓴다고 생각하면 쉬움.
- 첫번째 인자 : 함수명
- 두번째 인자 : 순서가 있는 자료형

````python
# map(함수명, 순서가 있는 자료형)
mylist = [1,2,3,4,5]
plus10 = lambda num : num+10

mylist = list(map(plus10, mylist))
print(mylist)

#----------------------------------
# map 안에도 lambda를 쓸 수 있다.
mylist = [1,2,3,4,5]
mylist = list(map(lambda num: num+10, mylist))
print(mylist)

#----------------------------------
# 튜플도 된다.

mylist = (1,2,3,4,5)
mylist = list(map(lambda num: num+10, mylist))     # 튜플로 만들었지만 내가 여기서 리스트로 바꿨다.
print(mylist)

#----------------------------------
# set도 가능.
mylist = [1,2,3,4,5]
mylist = set(map(lambda num: num+10, mylist))
print(mylist)
print(map(lambda num: num+10, mylist))
````

### ③ 재귀함수()
- 함수 내에서 그 함수 스스로를 호출하게 만드는 방식.

````python
def hello(count):
    if count == 0:  # 종료조건 만듦. count가 0이면 다시 hello 함수 호출하지 않고 끝냄.
        return
    
    print('Hello, world!', count)

    count -= 1      # count를 1 감소시킨 뒤
    hello(count)    # 다시 hello에 넣음

hello(5)            # hello 함수 호출
````

#### 예제. 팩토리얼 재귀함수

````python
# 팩토리얼을 재귀함수로 만들기

def factorial(n):
    if n == 1:                    # n이 1일 때
        return 1                  # 1을 반환하고 재귀 호출을 끝냄
    print(n)
    return n * factorial(n - 1)   # n과 factorial 함수에 n-1 을 넣어서 반환된 값을 곱함

print( factorial( int( input('팩토리얼 구하려는 숫자: ') ) ) )
````
