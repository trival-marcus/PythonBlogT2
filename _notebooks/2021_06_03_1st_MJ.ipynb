## 예제. DT스쿨 최저 연령은?

```python
# 선생님이 보내준 코드
# 간단한 파일 형식으로 쓰기 좋다.

alist = {
    '1팀':[25,39,29,27,22], 
    '2팀':[27,24,25,29,25], 
    '3팀':[25,26,25,23,23,28,27],
    '4팀':[21,23,30],
    '5팀':[26,41,24,31,21,34,27],
    '6팀':[21,29,27,25,25]}
age_li = []
for i,v in alist.items():
    print(f'{i}의 최저연령은{min(v)} 최고연령은 {max(v)} 평균연령은 {v / len(alist.values()}')
                        # 기억안남
    age_li.extend(v)
    # print(age_li)
print(f'DT스쿨 최저연령은{min(age_li)} 최고연령은 {max(age_li)}')
```
